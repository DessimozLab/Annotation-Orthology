```python
import seaborn as sns
import scipy as sp

sns.set_style("white")
fig, ax = plt.subplots(figsize=(20,12))
ax1=sns.scatterplot(data=abin, x='Divergence-time', y='count', hue='Parameters species', s=80)
for i, txt in enumerate(abin['Species']):
    xpos= abin['Divergence-time'][i]
    ypos=abin['count'][i]
    fs=13.5
    if txt =="D. rerio (zebrafish)":
        ax.annotate(txt, (xpos -5, ypos+1200), fontsize=fs)
    elif txt== "M. musculus (house mouse)":
        ax.annotate(txt, (xpos -10, ypos+1600), fontsize=fs)
    elif txt == "T. guttata (zebra finch)":
        ax.annotate(txt, (xpos-5, ypos-1800), fontsize=fs)
        
    elif txt == "S. habroptila (kakapo)":
        ax.annotate(txt, (xpos+4, ypos), fontsize=fs)
    else:
        ax.annotate(txt, (xpos + 4, ypos-800), fontsize=fs)

# Pearson's correlation
rp, pp = sp.stats.pearsonr(x=abin['Divergence-time'], y=abin['count'])


## Correlation line
import numpy as np
m, b = np.polyfit(abin['Divergence-time'], abin['count'], 1)
X_plot = np.linspace(ax.get_xlim()[0],ax.get_xlim()[1],100)
plt.plot(X_plot, m*X_plot + b, '-', c='pink')

plt.annotate('r'' = {:.2f}'.format(rp) + ', p = {:.3f}'.format(pp), xy=(0.25, 0.4), 
             xycoords='axes fraction', fontsize=22)


ax.set_xlim(-10,450)
sns.despine()
ax.set_ylabel('Number of proteins', labelpad =22, size =20)
ax.set_xlabel('Approximate divergence time (MYa) with parameters species', labelpad=22, size=20)
plt.legend(bbox_to_anchor=(0.05, 0.85), loc='upper left', fontsize=18,
           title= "Parameters species", title_fontsize=18)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.savefig('Augustus-Nproteins-divergencetime_correl.png', bbox_inches="tight", dpi=300)
```
